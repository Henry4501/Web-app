```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Define the original square
square = np.array([
    [0, 0],
    [0, 3],
    [3, 3],
    [3, 0],
    [0, 0]  # Close the shape
])

# Step 2: Define scaling matrix (scale x by 2, y by 1)
scaling = np.array([
    [2, 0],
    [0, 1]
])

# Step 3: Define rotation matrix (rotate by 45 degrees)
theta_deg = 45
theta_rad = np.radians(theta_deg)
rotation = np.array([
    [np.cos(theta_rad), -np.sin(theta_rad)],
    [np.sin(theta_rad),  np.cos(theta_rad)]
])

# Step 4: Combine scaling and rotation
combined_transformation = rotation @ scaling

# Step 5: Apply transformation to each point
transformed_square = square @ combined_transformation.T

# Step 6: Plotting both shapes
plt.figure(figsize=(8, 8))
plt.plot(square[:, 0], square[:, 1], 'bo-', label='Original Square')
plt.plot(transformed_square[:, 0], transformed_square[:, 1], 'go-', label='Scaled + Rotated')
plt.title('Scaling + Rotation Transformation in 2D')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.axis('equal')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend()
plt.show()

```
